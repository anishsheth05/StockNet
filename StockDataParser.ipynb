```python
import pandas as pd
import numpy as np
import pickle as pck
import time as tm
```

```python
with open('stock_data.pickle', 'rb') as fileloc:
    raw = pck.load(fileloc)
```

```python
data = np.ndarray((3721), dtype=pd.DataFrame)
for i in range(len(data)):
    data[i] = raw[i]
```

```python
hope = 0
hun_hope = 0
thou_hope = 0
yeet = 0
for frame in data:
    if len(frame) > 11000:
        yeet += 1
    elif len(frame) > 1100:
        thou_hope += 1
    elif len(frame) > 110:
        hun_hope += 1
    elif len(frame) > 11:
        hope += 1
print('Minimum of min requirement:', hope, '\nEven better at hundred greater:', hun_hope, '\nBut only the best at thousand plus:', thou_hope, '\nYeet:', yeet)
print(1000*yeet+100*thou_hope+10*hun_hope+hope)
```

```python
parsed = []
start = tm.time()

for stock in data:
    
    sample_num = 0
    
    while len(stock.Close) >= (sample_num+1)*11:
        
        inputs = np.full((40), np.nan)
        
        for i in range(10):
            
            inputs[(i*4)] = stock.Open[(sample_num*11)+i]
            inputs[(i*4)+1] = stock.Close[(sample_num*11)+i]
            inputs[(i*4)+2] = stock.High[(sample_num*11)+i]
            inputs[(i*4)+3] = stock.Low[(sample_num*11)+i]
            
        output_diff = stock.Close[(sample_num*11)+10] - stock.Close[(sample_num*11)+9]
        if stock.Close[(sample_num*11)+9] == stock.Close[(sample_num*11)+10]:
            output_per = 0
        elif stock.Close[(sample_num*11)+9] == 0:
            output_per = np.nan
        else:
            output_per = output_diff*100/stock.Close[(sample_num*11)+9]
        if abs(output_per) > 1:
            output_bin = np.sign(output_diff)
        else:
            output_bin = 0
        outputs = np.array([output_diff, output_per, output_bin])
        
        sample = np.array([inputs, outputs])
        if (not (np.isnan(inputs).any())) and (not (np.isnan(outputs).any())):
            parsed += [sample]
        if float(len(parsed))%100000==0 and len(parsed) > 0:
            print('Hundred thousand done!')
        elif float(len(parsed))%10000==0 and len(parsed) > 0:
            print('Ten thousand done.')
        sample_num += 1
parsed = np.asarray(parsed)

end = tm.time()
stopwatch = end - start
print('Process took', stopwatch, 'seconds.')
```

```python
# getting started with two variables with the highest scope: one being the list where the samples are contained,
# the other variable containing the start time of the parsing
parsed = []
start = tm.time()

# loops through every single stock's data, with the stock's data contained in the variable 'stock'
for stock in data:
    
    sample_num = 0
    
    while len(stock.Close) >= (sample_num+1)*11:
        
        inputs = np.full((40), np.nan)
        current = stock.iloc[sample_num*11:(sample_num+1)*11, :4]
        current = (current - current.Low.min()) / (current.High.max() - current.Low.min())
        
        for i in range(10):
            
            inputs[(i*4)] = current.Open[i]
            inputs[(i*4)+1] = current.Close[i]
            inputs[(i*4)+2] = current.High[i]
            inputs[(i*4)+3] = current.Low[i]
            
        output_diff = current.Close[10] - current.Close[9]
        if abs(output_diff) > 0.10:
            output_bin = np.sign(output_diff)
        else:
            output_bin = 0
        output = np.array([output_diff, output_bin])
        
        sample = np.array([inputs, output])
        if (not (np.isnan(inputs).any())) and (not (np.isnan(output).any())):
            parsed += [sample]
            
        if float(len(parsed))%100000==0 and len(parsed) > 0:
            print('Hundred thousand done!')
        elif float(len(parsed))%10000==0 and len(parsed) > 0:
            print('Ten thousand done.')
        sample_num += 1

parsed = np.asarray(parsed)

end = tm.time()
stopwatch = end - start
print('Process took', stopwatch, 'seconds.')
```

```python
# this time we will be smart and parse data into both input and output separately - better for numpy
parsed_inputs = []
parsed_outputs = []

# loops through every single stock's data, with the stock's data contained in the variable 'stock'
for stock in data:
    
    sample_num = 0
    
    while len(stock.Close) >= (sample_num+1)*11:
        
        inputs = []
        current = stock.iloc[sample_num*11:(sample_num+1)*11, :4]
        current = (current - current.Low.min()) / (current.High.max() - current.Low.min())
        
        for i in range(10):
            inputs.append(current.Open[i])
            inputs.append(current.Close[i])
            inputs.append(current.High[i])
            inputs.append(current.Low[i])
            
        output_diff = current.Close[10] - current.Close[9]
        if abs(output_diff) > 0.10:
            output_bin = np.sign(output_diff)
        else:
            output_bin = 0
        output = np.array([output_bin, output_diff])
        
        
        if (not (np.isnan(inputs).any())) and (not (np.isnan(output).any())):
            parsed_inputs.append(inputs)
            parsed_outputs.append(output)
            
        if float(len(parsed_inputs))%100000==0 and len(parsed_inputs) > 0:
            print('Hundred thousand done!')
        elif float(len(parsed_inputs))%10000==0 and len(parsed_inputs) > 0:
            print('Ten thousand done.')
        sample_num += 1

parsed_inputs = np.asarray(parsed_inputs)
parsed_outputs = np.asarray(parsed_outputs)
```

```python
with open('parsed_data_inputs.pickle', 'wb') as fileloc:
    pck.dump(parsed_inputs, fileloc)
with open('parsed_data_outputs.pickle', 'wb') as fileloc:
    pck.dump(parsed_outputs, fileloc)
```
